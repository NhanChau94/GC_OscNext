```python
#############################################################################################################
#   Create .dag and submit script
#
submit = f'{submit_dir}/submit.sh'
dag_file = '{}/TS.dag'.format(submit_dir)


#list of channels and profiles
# channels = ["numunumu"]
channels = ["WW","bb","tautau"]
# profile = ['NFW']
profile = ['NFW','Burkert']


# process
prcs = ["ann", "decay"]

# mass
mass = [100, 500, 1000] #error1, error2

# number of mass scan point
n = 30


# bkg
bkgs = ['FFT', 'sklearn', 'nokde']


samples = ['null_hypothesis', 'RA_scramble']

# Loops on scenarios and submit jobs for each of them
with open(dag_file, 'w') as f:
    for c in channels:
        for p in profile:
            for prc in prcs:
                for m in mass:
                    for bkg in bkgs:
                        for sample in samples:    
                            output = f'/data/user/tchau/DarkMatter_OscNext/TS_distribution/{c}_{p}_{prc}_{m}_bkg_{bkg}_sample_{sample}_test.pkl'
                            f.write('JOB TS_{}_{}_{}_{}_{}_{}'.format(prc, c, p, m, bkg, sample)+' TS.submit \n')
                            f.write('VARS TS_{}_{}_{}_{}_{}_{}'.format(prc, c, p, m, bkg, sample)
                                    +' JOBNAME="TS_{}_{}_{}_{}_{}_{}"'.format(prc, c, p, m, bkg, sample) 
                                    +' c="{}" p="{}" proc="{}" mass="{}" bkg="{}" sample="{}" outputpath="{}"'.format(c, p, prc, m, bkg, sample, output)+'\n')
f.close()
with open(submit, 'w') as s:
    s.write("#!/bin/sh\n")
    s.write("condor_submit_dag -config dagman.config {}\n".format(dag_file))
    print("condor_submit_dag -config dagman.config {}\n".format(dag_file))
s.close()
```

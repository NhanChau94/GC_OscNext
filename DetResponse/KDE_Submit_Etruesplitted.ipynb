```python

cfig_file = '/home/tchau/code/condor_submit/Etruesplitted/dagman.config'
with open(cfig_file, 'w') as c:
    c.write('DAGMAN_MAX_JOBS_SUBMITTED=500'+'\n')
    c.write('DAGMAN_MAX_SUBMIT_PER_INTERVAL=20'+'\n')
    c.write('DAGMAN_USER_LOG_SCAN_INTERVAL=10'+'\n')
    c.close()

sub_file = '/home/tchau/code/condor_submit/Etruesplitted/Resp_KDE.submit'
with open(sub_file, 'w') as s:
    s.write('executable = /data/user/tchau/Sandbox/GC_OscNext/DetResponse/KDE_RespMatrix_Etruesplitted.py'+'\n')
    s.write('\n')
    s.write('initialdir = /data/user/tchau/'+'\n')
    s.write('logdir = /scratch/tchau/DM_OscNext/MakePDF/logs/'+'\n')
    s.write('output = $(logdir)$(JOBNAME).$(Cluster).out'+'\n')
    s.write('error = $(logdir)$(JOBNAME).$(Cluster).err'+'\n')
    s.write('log = $(logdir)$(JOBNAME).$(Cluster).log'+'\n')
    s.write('\n')
    s.write('notification   = never'+'\n')
    s.write('universe       = vanilla'+'\n')
    s.write('\n')
    s.write('should_transfer_files = YES'+'\n')
    s.write('request_memory = 5GB'+'\n')
    s.write('\n')
    s.write('arguments = -b $(bandwidth) -m $(package) -n $(nu) -p $(pid) -f $(firstE) -l $(lastE)'+'\n')
    s.write('queue')
    s.close()

nutypes=["nu_e", "nu_mu", "nu_tau", "nu_e_bar", "nu_mu_bar", "nu_tau_bar"]
pids = [0,1,2]

# nutypes=["nu_mu"]
# pids = [2]
submit = '/home/tchau/code/condor_submit/Etruesplitted/submit.sh'

# Number of true E per each job:
N_Etrue = 2

with open(submit, 'w') as s:
    s.write("#!/bin/sh\n")
    for nu in nutypes:
        for pid in pids:
            path = '/home/tchau/code/condor_submit/Etruesplitted/{}_{}'.format(nu, pid)
            if not (os.path.exists(path)): os.makedirs(path)
            for Eid in range(0, 300-(N_Etrue-1), 2):
                firstE = Eid
                lastE = Eid + (N_Etrue-1)
                
                dag_file = '{}/RespKDE_{}_Bandwith_{}_{}_pid{}_Etrue_{}_{}.dag'.format(path, package, bandwidth, nu, pid, firstE, lastE)
                with open(dag_file, 'w') as f:
                    f.write('JOB RespKDE_{}_Bandwith_{}_{}_pid{}_Etrue_{}_{}'.format(package, bandwidth, nu, pid, firstE, lastE)+' Resp_KDE.submit \n')
                    f.write('VARS RespKDE_{}_Bandwith_{}_{}_pid{}_Etrue_{}_{}'.format(package, bandwidth, nu, pid,  firstE, lastE)+' JOBNAME="RespKDE_{}_Bandwith_{}_{}_pid{}_Etrue_{}_{}"'.format(package, bandwidth, nu, pid,  firstE, lastE) +' bandwidth="{}" package="{}" nu="{}" pid="{}" firstE="{}" lastE="{}"'.format(bandwidth, package, nu, pid, firstE, lastE)+'\n')
                f.close()
                s.write("condor_submit_dag -config dagman.config {} \n".format(dag_file))
s.close()
```
